### Research Question: 
What factors best predict an individualâ€™s wage using the `Wage` dataset? 

```python
import pandas as pyimport 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
%matplotlib inline
```

## Check wage dataset

```python
#convert csv file to panda dataframe and dictionary
wage_df = pd.read_csv('wage.csv')
wage_data = wage_df.to_dict()
```

```python
#check on first 10 ID
wage_df.head(10)
```

```python
len(wage_df)
```

## Data visualization

```python
#plot histogram chart for continuous variables
continuous_variables =['year','age','logwage','wage']
for i in list(range(0, len(continuous_variables))):
    list_data_for_the_variable = list(wage_data[continuous_variables[i]].values())
    plt.figure(figsize=(8,6))
    plt.title(continuous_variables[i])
    plt.hist(x = list_data_for_the_variable, bins = 5, edgecolor='black')
    plt.xlabel(continuous_variables[i])
    plt.ylabel('frequency')
```

```python
#plot histgram chart for categorical variables
categorical_variables =['maritl','race','education','region','jobclass','health','health_ins']
for i in list(range(0, len(categorical_variables))):
    list_values_for_the_variable = list(wage_data[categorical_variables[i]].values())   
    list_keys_for_the_variable = list(wage_data[categorical_variables[i]].keys())
    plt.figure(figsize=(8,6))
    plt.title(categorical_variables[i])
    plt.hist(x = list_values_for_the_variable, bins = 5, edgecolor='black')
    plt.xlabel(categorical_variables[i])
    plt.ylabel('numbers')
```

## Fit with a linear OLS regression model

```python
#Fit with OLS linear regression model
#model
result = smf.ols('wage ~ year + education + age + race + maritl + region + jobclass + health + health_ins + logwage', data=wage_df).fit()
print(result.summary())
```

### Visualization of how each variable affacts with wage

```python
fig = plt.figure(figsize=(20,18))
fig = sm.graphics.plot_partregress_grid(result, fig=fig)
```
