### While, break, and continue
What does a while loop look like in Python?

````python
i = 1
while i < 6:
  print(i)
  i += 1
````

- What is break and continue?
- Are they different? How?
- Run the following code:

````python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print("foo")
    continue
  print(i)
````

How does the code above work?

Now run this code:

````python
i = 0
while i < 6:
  i += 1
  if i == 3:
    print("foo")
    break
  print(i)
````
